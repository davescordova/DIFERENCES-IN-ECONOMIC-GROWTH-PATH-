```python
import pandas as pd

# Paths to the Excel files
file_path_geral = r"C:\Users\daves\OneDrive\Pessoal\Acadêmico\Mestrado\Dissertação - Execução\Dissertação\Quadros e gráficos\Comércio exterior\H_EXPORTACAO_GERAL_1997-01_2024-12_DT20250512.xlsx"
file_path_mt = r"C:\Users\daves\OneDrive\Pessoal\Acadêmico\Mestrado\Dissertação - Execução\Dissertação\Quadros e gráficos\Comércio exterior\H_EXPORTACAO_POR_MUNICIPIO_1997-01_2024-12_DT20250512.xlsx"

geral = pd.read_excel(file_path_geral, sheet_name="Resultado")
mt = pd.read_excel(file_path_mt, sheet_name="Resultado")
```

```python
import matplotlib.pyplot as plt


# Define the list of value columns in ascending order of year
value_columns_ordered = [
    '1997 - Valor US$ FOB', '1998 - Valor US$ FOB', '1999 - Valor US$ FOB', '2000 - Valor US$ FOB',
    '2001 - Valor US$ FOB', '2002 - Valor US$ FOB', '2003 - Valor US$ FOB', '2004 - Valor US$ FOB',
    '2005 - Valor US$ FOB', '2006 - Valor US$ FOB', '2007 - Valor US$ FOB', '2008 - Valor US$ FOB',
    '2009 - Valor US$ FOB', '2010 - Valor US$ FOB', '2011 - Valor US$ FOB', '2012 - Valor US$ FOB',
    '2013 - Valor US$ FOB', '2014 - Valor US$ FOB', '2015 - Valor US$ FOB', '2016 - Valor US$ FOB',
    '2017 - Valor US$ FOB', '2018 - Valor US$ FOB', '2019 - Valor US$ FOB', '2020 - Valor US$ FOB',
    '2021 - Valor US$ FOB', '2022 - Valor US$ FOB', '2023 - Valor US$ FOB', '2024 - Valor US$ FOB'
]

if 'País' in mt.columns and '2024 - Valor US$ FOB' in mt.columns and 'Países' in geral.columns:
    # Calculate the total export value for each country in 2024 in 'mt' DataFrame
    country_2024_exports_mt = mt.groupby('País')['2024 - Valor US$ FOB'].sum()

    # Get the top 10 countries with the highest export value in 2024 from 'mt'
    top_10_countries = country_2024_exports_mt.nlargest(10).index.tolist()

    # Filter the 'mt' DataFrame for the top 10 countries
    mt_top_10 = mt[mt['País'].isin(top_10_countries)].copy()

    # Calculate the total amount for each year in 'mt' (for the top 10)
    total_amounts_top_10_mt = mt_top_10[value_columns_ordered].sum()

    # Calculate percentage participation of top 10 countries in 'mt'
    plt.figure(figsize=(15, 8))
    num_years = len(value_columns_ordered)
    x_positions = range(num_years)

    for country in top_10_countries:
        country_data_mt = mt_top_10[mt_top_10['País'] == country]
        country_total_per_year_mt = country_data_mt[value_columns_ordered].sum()
        country_percentages_mt = [(country_total / total) * 100 if total != 0 else 0
                                  for country_total, total in zip(country_total_per_year_mt, total_amounts_top_10_mt)]
        plt.plot(x_positions, country_percentages_mt, label=country)

    # --- Plotting China from 'geral' DataFrame ---
    if 'China' in geral['Países'].unique():
        china_data_geral = geral[geral['Países'] == 'China'][value_columns_ordered].sum()
        total_amounts_geral = geral[value_columns_ordered].sum()
        china_percentages_geral = [(china_total / total) * 100 if total != 0 else 0
                                   for china_total, total in zip(china_data_geral, total_amounts_geral)]
        plt.plot(x_positions, china_percentages_geral, label='China (Geral)', linestyle='--', linewidth=2, color='black')
    else:
        print("Warning: 'China' not found in the 'Países' column of the 'geral' DataFrame.")

    plt.xlabel("Year")
    plt.ylabel("Percentage of Total (%)")
    plt.xticks(x_positions, [col.split(' - ')[0] for col in value_columns_ordered], rotation=45, ha="right")
    plt.legend(title="Countries", loc='upper center', bbox_to_anchor=(0.5, -0.15),
               fancybox=True, shadow=True, ncol=3)
    plt.tight_layout()
    plt.grid(True)


    # Save the plot
    output_path = r"C:\Users\daves\OneDrive\Pessoal\Acadêmico\Mestrado\Dissertação - Execução\Dissertação\Quadros e gráficos\Comércio exterior\export_plot.jpg"
    plt.savefig(output_path, dpi=1200, format='jpg')
    print(f"Plot saved to: {output_path} with DPI 1200")

    plt.show()

else:
    error_messages = []
    if 'País' not in mt.columns:
        error_messages.append("'País' column not found in the 'mt' DataFrame.")
    if '2024 - Valor US$ FOB' not in mt.columns:
        error_messages.append("'2024 - Valor US$ FOB' column not found in the 'mt' DataFrame.")
    if 'Países' not in geral.columns:
        error_messages.append("'Países' column not found in the 'geral' DataFrame.")
    print("Error:", " ".join(error_messages))
    print("Please ensure the column names are correct.")
```
